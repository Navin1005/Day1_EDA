```python
# Scenario 1: Analyzing Trends Over Time
#Graph to Use: Line Plot
# Why: Shows how a variable changes over time (e.g., sales, temperature).
import pandas as pd
import matplotlib.pyplot as plt

# Example dataset: Monthly sales
data = {'Month': ['Jan', 'Feb', 'Mar', 'Apr'],
        'Sales': [200, 250, 300, 400]}
df = pd.DataFrame(data)

# Line Plot
plt.plot(df['Month'], df['Sales'], marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
#Scenario 2: Comparing Categories
#Graph to Use: Bar Plot
#Why: Displays differences between categories (e.g., sales by region or product type).
import seaborn as sns

# Example dataset
data = {'Product': ['A', 'B', 'C'],
        'Sales': [100, 150, 200]}
df = pd.DataFrame(data)

# Bar Plot
sns.barplot(x='Product', y='Sales', data=df)
plt.title("Sales by Product")
plt.show()
```

```python
#Scenario 3: Distribution of a Single Variable
#Graph to Use: Histogram or KDE Plot
#Why: Reveals the frequency distribution of data (e.g., customer age, transaction amount).
import seaborn as sns

# Example dataset
data = [22, 23, 45, 22, 34, 45, 36, 28, 29, 30]

# Histogram
sns.histplot(data, bins=5, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
#Scenario 4: Analyzing Relationships Between Two Variables
#Graph to Use: Scatter Plot
#Why: Shows the correlation or pattern between two variables (e.g., advertising spend vs sales).
```

```python
import seaborn as sns

# Example dataset
data = {'Ad_Spend': [100, 200, 300, 400],
        'Sales': [10, 20, 25, 30]}
df = pd.DataFrame(data)

# Scatter Plot
sns.scatterplot(x='Ad_Spend', y='Sales', data=df)
plt.title("Ad Spend vs Sales")
plt.show()
```

```python
# Scenario 5: Understanding Feature Correlation
# Graph to Use: Heatmap
# Why: Highlights the correlation between variables in a dataset (e.g., which features influence sales).
import seaborn as sns
import pandas as pd

# Example dataset
data = {'Feature1': [1, 2, 3, 4],
        'Feature2': [4, 3, 2, 1],
        'Sales': [10, 20, 30, 40]}
df = pd.DataFrame(data)

# Correlation Heatmap
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Feature Correlation")
plt.show()
```

```python
# Scenario 6: Analyzing Outliers
# Graph to Use: Box Plot
# Why: Displays the spread of data and identifies outliers (e.g., transaction amounts, delivery times).
import seaborn as sns

# Example dataset
data = {'Delivery_Time': [5, 7, 9, 6, 5, 20, 8]}
sns.boxplot(data=data)
plt.title("Delivery Time Analysis")
plt.show()
```

```python
# Scenario 7: Comparing Distributions
# Graph to Use: Violin Plot
# Why: Shows both the distribution and probability density of data for categories (e.g., exam scores by gender).
import seaborn as sns

# Example dataset
data = {'Category': ['A', 'A', 'B', 'B', 'C', 'C'],
        'Values': [10, 12, 15, 20, 18, 25]}
df = pd.DataFrame(data)

sns.violinplot(x='Category', y='Values', data=df)
plt.title("Category Distribution")
plt.show()
```

```python
# Scenario 8: Multi-Variable Relationships
# Graph to Use: Pair Plot
#Why: Visualizes pairwise relationships for multiple variables in a dataset (e.g., Iris dataset).
import seaborn as sns
import pandas as pd

# Load dataset
df = sns.load_dataset('iris')
sns.pairplot(df, hue='species')
plt.title("Pairwise Relationships")
plt.show()
```

```python
# Scenario 9: Showing Proportions
# Graph to Use: Pie Chart (Matplotlib) or Count Plot (Seaborn)
# Why: Shows percentage distribution (e.g., customer demographics).
import matplotlib.pyplot as plt

labels = ['Male', 'Female', 'Others']
sizes = [50, 45, 5]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Customer Demographics")
plt.show()
```

```python
import seaborn as sns

data = {'Gender': ['Male', 'Female', 'Male', 'Female', 'Others']}
df = pd.DataFrame(data)

sns.countplot(x='Gender', data=df)
plt.title("Customer Demographics")
plt.show()
```
